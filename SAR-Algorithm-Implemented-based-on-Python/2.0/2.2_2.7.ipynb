```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as ss
```

线性卷积
    一维卷积

```python
x = np.array([1,3,-1,5,2,6,4,-2])
h = np.array([1,2,3])
print(ss.convolve(x,h))
```

线性卷积
    二维卷积

```python
h_1 = np.array([1,3,1]).reshape(3,1)
h_2 = np.array([1,2,2,1]).reshape(1,4)
h = h_1@h_2
print('h:',h)
x = np.ones([4,4])
print(ss.convolve2d(x,h))
print(ss.convolve(ss.convolve(x,h_1),h_2))
```

sinc插值拟合

```python
g_d = np.array([1,4,2,6,5,-4.5,3,4.5])
L = np.array(range(8))
plt.figure()
plt.grid()
plt.plot(L,g_d,'*')
g_tar = np.sum(np.sinc(3.7-L)*g_d)/(np.sum(np.sinc(3.7-L)))
plt.plot(3.7,g_tar,'o')
print(np.sinc(3.7-L)*g_d)
```
