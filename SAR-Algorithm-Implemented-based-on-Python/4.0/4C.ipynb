```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as ss
import scipy.fftpack as sf
```

天线波束宽度

```python
N = np.array([3,7,15,31,63,127,1023])
theta = np.linspace(-np.pi/2,np.pi/2,4001).reshape(1,4001)
L = 10
lam = 3e-1
for i in range(len(N)):
    x_n = np.linspace(-L/2,L/2,N[i]).reshape(N[i],1)
    P = np.exp(-1j*2*np.pi*np.kron(x_n,np.sin(theta))/lam)
    P_sum = np.sum(P,axis=0)*L/N[i]
    P_approx =  L*np.sinc(L*theta/lam)
    plt.figure()
    plt.plot(theta.flatten(),np.abs(P_sum))
    plt.plot(theta.flatten(),np.abs(P_approx).flatten())
    plt.xlim(-np.pi/10,np.pi/10)
    
    
```
