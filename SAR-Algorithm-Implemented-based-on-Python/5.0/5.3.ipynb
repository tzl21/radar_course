```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as ss
import scipy.fftpack as sf
```

相位耦合

```python
TBP = 38
T = 7.2e-6
K = TBP/(T**2)

ADC_sampling_frequency = 256*K*T
interval_base = 1/ADC_sampling_frequency
N = int(T*ADC_sampling_frequency)
NN = 256

t_l = np.linspace(-T/2,T/2-interval_base,N)
t_c = np.linspace(0,-T/2,NN).reshape(NN,1)
x_chirp_2d = np.exp(1j*np.pi*K*((t_l.reshape(1,N)-t_c)**2))
print('TBP:',round(K*T**2))
print('N:',N)

plt.figure()
plt.subplot(3,1,1)
plt.plot(t_l*1e+6,np.real(x_chirp_2d[0,:]))
plt.subplot(3,1,2)
plt.plot(t_l*1e+6,np.real(x_chirp_2d[10,:]))
plt.subplot(3,1,3)
plt.plot(t_l*1e+6,np.real(x_chirp_2d[20,:]))
plt.tight_layout()

x_chirp_v = x_chirp_2d[:,int(N/2)-1000]
plt.figure()
plt.plot(np.real(x_chirp_v),range(NN))
```
